#### parameterized functions without return value

```
def fibonacci(size):
    fib = [0,1]
    for i in range(size):
        val = fib[-2] + fib[-1]
        fib.append(val)
    print(fib)
```

```
fibonacci(5)
```

```
fibonacci(15)
```

```
print(fibonacci(5))
```

#### parameterized function with return value

```
def fibonacci2(size):
    fib = [0,1]
    for i in range(size):
        val = fib[-2] + fib[-1]
        fib.append(val)
    return fib
```

```
fibonacci2(5)
```

```
print(fibonacci2(5))
```

```
result1 = fibonacci(10)
result2 = fibonacci2(10)
```

```
print(result1)
```

```
print(result2)
```

```
def word_reverse(data):
    words = data.split()
    reversed_data = []
    for item in words:
        rword = item[::-1]
        reversed_data.append(rword)
    return " ".join(reversed_data)
```

```
word_reverse('hello how are you')
```

```
word_reverse('there was the biggest pizza, that i saw')
```
