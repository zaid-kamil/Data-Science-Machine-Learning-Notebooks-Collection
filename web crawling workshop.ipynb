# Basic web crawler example
libraries that we need
- requests
- beautifulsoup

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://timesofindia.indiatimes.com/home/headlines"
```

```python
page =  requests.get(url) # raw
```

```python
print(page)
```

```python
soup = BeautifulSoup(page.text)
```

```python
print(soup.title)
```

```python
links = soup.find_all('a') # find all those tags that are 'a' and give me the list
```

```python
for item in links:
    print(item.text)
```
