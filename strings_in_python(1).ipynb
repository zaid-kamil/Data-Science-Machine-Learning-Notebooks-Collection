```
url = input('enter a url of image or doc or pdf:')
if url.endswith('.png') or url.endswith('.jpg'):
    print('its a image u want to download')
elif url.endswith('.docx'):
    print("you want to downlaod a document")
elif url.endswith('.pdf'):
    print('you want to download a PDF')
else:
    print('unknown file type')
```

### wap a program to remove all vowels from a string 
input - i want icecream

output - wnt ccrm

vowel are removed

hint=> use a for loop


```
vowels = "aeiou"
data = 'the way to learn machine learning is to practice data science'
for v in vowels :
    data = data.replace(v,'')
print(data)
```
