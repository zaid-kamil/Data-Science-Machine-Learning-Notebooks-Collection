```
years = np.arange(2010,2021)
print(years, len(years))
rainfall = np.random.randint(100,350,11)
print(rainfall, len(rainfall))
```

```
plt.pie(rainfall, labels=years)
plt.show()
```

```
plt.pie(rainfall, labels=years, autopct='%.1f', radius=2, shadow=True)
plt.show()
```

```
plt.pie(rainfall, labels=years, autopct='%.1f', radius=2, shadow=True, wedgeprops={'width':.7})
plt.show()
```

```
plt.bar(years, rainfall)
```

```
plt.bar(years, rainfall,color=['#89f','#f9f'])
```

```
plt.boxplot(rainfall)
plt.show()
```
