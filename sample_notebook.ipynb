```python
model = tf.keras.Sequential()
model
```

```python
model.add(tf.keras.layers.Dense(units=5,
    activation = tf.keras.activations.relu,
    input_shape = (10,))
)
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
model.summary()
```
