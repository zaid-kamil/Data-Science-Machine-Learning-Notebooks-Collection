there are five was to display outout using print function

```python
a = 10
b = "apples"
price = 300
```

```python
# easiest option (use comma)
print('i purchased', a, b, 'for rupees', price)
```

```python
# c style
print('i purchased %d %s for rupees %d'%(a,b,price))
```

```python
# java style
print("i purchased "+str(a)+' '+b+' for rupees'+str(price))
```

```python
# python format function
print('i purchased {} {} for rupees {}'.format(a,b,price))
```

```python
# python f string (>3.6)
print(f'i puchased {a} {b} for rupees {price}')
```
