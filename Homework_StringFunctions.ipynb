```python
# wap to replace all special characters with _

student = 'Krishna is a good boy.'
R = student.replace('good','_')
print(R)
```

```python
# wap to remove all vowels in a string.

sentence = "Proud to be an Indian"

for v in 'aeiou' :
  sentence = sentence.replace(v,' ')
print(sentence)
```

```python
# wap to count all the characters in a string.

s2 = "you cannot change your future but you can change your habits and surely your habits will change your future"

A = s2.count('you')               # your is also include as it contains you
print(A)
B = s2.count('cannot')
print(B)
C = s2.count('change')
print(C)
D = s2.count('your')
print(D)
E = s2.count('future')
print(E)
F = s2.count('but')
print(F)
G = s2.count('can')         # cannot is also included as it contains can
print(G)
H = s2.count('habits')
print(H)
I = s2.count('and')
print(I)
J = s2.count('surely')
print(J)
K = s2.count('will')
print(K)
L = s2.count(' ')
print(L)
```

```python
# WAP TO COUNT ALL THE WORDS IN A STRING

word = "Life is very precious, All is well, Life is a challenge"
A = word.count('Life')
print(A)

B = word.count('is')
print(B)

C = word.count('very')
print(C)

D = word.count('precious')
print(D)

E = word.count('All')
print(E)

G = word.count('well')
print(G)

H = word.count('a')
print(H)

I = word.count('challenge')
print(I)
```
