```
imgs_hold = "/content/drive/My Drive/gesture_detection_system/hold.zip"
imgs_none = "/content/drive/My Drive/gesture_detection_system/none.zip"
imgs_play = "/content/drive/My Drive/gesture_detection_system/play.zip"
```

```
from zipfile import ZipFile 
import os
import cv2
from google.colab.patches import cv2_imshow
```

```
files = [imgs_hold,imgs_none,imgs_play]
folders =['hold','none','play']
path = '/content/drive/My Drive/gesture_detection_system/'
for file_name,folder_name in zip(files,folders):
    with ZipFile(file_name, 'r') as z:
        extract_path = path+"/"+folder_name 
        if not os.path.exists(extract_path):
            os.chdir(path)
            os.mkdir(folder_name)
        os.chdir(extract_path)
        z.extractall() 
```

```
def convert_image(image,folder):
    img = cv2.imread(image)
    bg =  cv2.createBackgroundSubtractorKNN()
    bgmask = bg.apply(img)
    print(os.getcwd())
    cv2.imwrite(image, bgmask)
    cv2_imshow(img)
    print(f'saved image to {folder}')
```

```
os.chdir(path)
for folder in folders:
    fpath = os.path.join(path,folder,)
    files = os.listdir(fpath)
    files = [os.path.join(fpath,img) for img in files]
    bgfolder = "bg_"+folder
    if not os.path.exists(bgfolder):
        os.mkdir(bgfolder)
    for img in files:
        convert_image(img, bgfolder)    
```
