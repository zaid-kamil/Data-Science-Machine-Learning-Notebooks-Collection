```
def clean_data(datalist):
    cleanlist = []
    for item in datalist:
        if isinstance(item, int):
            cleanlist.append(float(item))
        if isinstance(item, str):
            cleanlist.append(0.0)
    return cleanlist
```

```
x = [1,2,3,"a","nan",3]
```

```
clean_data(x)
```
