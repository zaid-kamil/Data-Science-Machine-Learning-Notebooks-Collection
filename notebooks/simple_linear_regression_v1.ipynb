```python
# basic libraries
import pandas as pd
import numpy as np
import plotly.express as px
# ml libraries
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/digipodium/Datasets/main/carsspeed.csv',usecols=['Hours','Scores'])
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
px.scatter(df, x='dist',y='speed')
```

```python
X = df[['dist']] # input can be one or more columns so we make it a 2 dimension matrix 
y = df['speed'] # output is always going to be a 1 dimension matrix (single column)
print(f"X = {X.shape},y= {y.shape}")
xtrain, xtest, ytrain, ytest  = train_test_split(X,y,test_size=.25, random_state=2)
print(f"xtrain = {xtrain.shape}, xtest =  {xtest.shape},\nytrain =  {ytrain.shape}, ytest = {ytest.shape}")
```

```python
model = LinearRegression()
model.fit(xtrain,ytrain)
print("trained the model")
```

```python
print("intercept")
model.intercept_
```

```python
print("coeficient")
model.coef_
```

```python
model.score(xtest,ytest) * 100
```

```python
model.score(xtest,ytest) * 100
```
