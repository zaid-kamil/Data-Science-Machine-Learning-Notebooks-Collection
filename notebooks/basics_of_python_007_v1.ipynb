# user defined functions
- syntax 1
```
def func_name(params...) :
    code line 1
    code line 2
    ...
    code line n
```
- syntax 2
```
def func_name(params...):
    code lines
    ...
    return value
```

userdefined function work in two stages
1. user writes the code of function (defination)
2. user uses the function by its name (calling) 

````
# $5,000.00 -> 5000
def clean_price(data):
    data = data[1:].replace(',','')
    data = float(data)
    return data
````

````
clean_price(data='$213,000.00')
````

````
product_prices = [
    '$29032',
    '$23.99',
    '$50,000.0',
    '$239,232'
]
````

````
for price in product_prices:
    print(clean_price(price)
````

# using functions in data analysis
- we have two ways of using a user defined functions on a dataset
1. use `apply`() to use a function on a single column
2. use `applymap`() to use a function on whole data
