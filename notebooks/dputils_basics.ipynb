```python
import pandas as pd
from dputils.scrape import Scraper, Tag
url = 'https://digipodium.com/mining-demo.html'
scraper = Scraper(url)
result = scraper.get_repeating_page_data(
    target = Tag(cls='item-list'),
    items = Tag(cls='col'),
    title = Tag('h4'),
    price = Tag('h1', cls='s-price'),
    discount = Tag('li', cls='discount'),
    oprice = Tag('li', cls='o-price'),
    bought = Tag('li', cls='total-bought'),
    left = Tag('li', cls='left'),
)
df = pd.DataFrame(result)
df
```

```python
df.price = df.price.apply(lambda i: float(i[1:]))
df.discount = df.discount.apply(lambda i: int(i[:2]))
```
