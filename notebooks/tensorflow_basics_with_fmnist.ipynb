# Import libraries

````
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow import keras
````

# download dataset

Basic classification: Classify images of clothing

Import the [Fashion MNIST dataset](https://github.com/zalandoresearch/fashion-mnist). The Fashion MNIST dataset which contains 70,000 grayscale images in 10 categories. The images show individual articles of clothing at low resolution (28 by 28 pixels)

````
data = keras.datasets.mnist
````

````
data = keras.datasets.fashion_mnist
````

````
(train_images,train_labels),(test_images,test_labels) = data.load_data()
````

# Explore image data

````
class_names=['T-shirt/Top','Trouser','Pullover','Dress','Coat',
             'Sandal','Shirt','Sneaker','Bag','Ankle boot']
````

# Normalize the data
**Scale the values to 0 and 1 range.**
```
value = value / maxvalue_in_series
```

````
train_images = train_images/255.0
test_images = test_images/255.0
````

````
print(train_images[4])
````

# Create a neural network layer list
using keras api


*   one flatten layer to flatten 28x28
*   one hidden layer of 128 neurons
*   one output layer with 10 neurons



````
# no convulation this time, data is prepared for direct use
layers = [keras.layers.Flatten(input_shape=(28,28)),
          keras.layers.Dense(128,activation='relu'),
          keras.layers.Dense(10,activation='softmax')]
````

# create a nueral network
Before the model is ready for training, it needs a few more settings. These are added during the model's compile step:

- **[Loss function](https://github.com/keras-team/keras/blob/master/keras/losses.py)** —This measures how accurate the model is during training. You want to minimize this function to "steer" the model in the right direction.
- **[Optimizer](https://towardsdatascience.com/types-of-optimization-algorithms-used-in-neural-networks-and-ways-to-optimize-gradient-95ae5d39529f)** —This is how the model is updated based on the data it sees and its loss function.
- **[Metrics](https://keras.io/metrics/)** —Used to monitor the training and testing steps. The following example uses accuracy, the fraction of the images that are correctly classified.

````
model = keras.Sequential(layers)
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
````

# train the NN

````
model.fit(train_images,train_labels, epochs=5)
````

# Evaluate the model train data


````
test_loss,test_acc = model.evaluate(test_images,test_labels)
````

````
print('test accuracy',test_acc)
````

# checking predictions

````
prediction = model.predict(test_images)
````

````
print(class_names[np.argmax(prediction[0])])
````

````
plt.figure(figsize=(16,6))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.grid(False)
    plt.imshow(test_images[i],cmap=plt.cm.binary)
    plt.xlabel("actual :"+class_names[test_labels[i]])
    plt.title("prediction :"+class_names[np.argmax(prediction[i])])
    plt.axis(False)
````
