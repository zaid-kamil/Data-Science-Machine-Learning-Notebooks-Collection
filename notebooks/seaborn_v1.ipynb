```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=sns.load_dataset('penguins')
df.head()
```

```python
sns.violinplot(x='island',y='body_mass_g',data=df, palette='Set2_r')
```

```python
sns.jointplot(x='bill_depth_mm',y='flipper_length_mm',data=df,kind='hex',color='#dd77ff')
plt.show()
```

```python
sns.jointplot(x='species',y='bill_length_mm',data=df,kind='scatter',color='blue')
plt.show()
```
