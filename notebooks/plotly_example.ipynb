```
import plotly.express as px
```

```
x = [1,2,3,4,5,6]
y = [2,4,6,6,3,3]
fig = px.scatter(x = x,y = y,width=800,size=x)
fig.show()
```

```
data = px.data.iris()
data.head()
```

```
fig = px.scatter(data,x='sepal_width',y='sepal_length')
fig.show()
```

```
fig = px.scatter_3d(data,x='sepal_width',y='sepal_length',z='petal_width',color='petal_width')
fig.show()
```
