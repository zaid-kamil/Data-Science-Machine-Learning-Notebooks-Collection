# print styles in python
- using commas
- using concatenation operator and cast functions
- using format specifiers
- using format function of string
- using f-string for shorter code

```
# using concatenation(+) operator
x = 10
y = 20
z = x + y
print(str(x) + "+" + str(y) + "="+ str(z)) # worst formatting option
```

```
# using format specifiers like %s, %d, %f
author = "brandon sanderson"
rank = 1
print("%s is on rank %d in authors" % (author, rank))
```

```
x = 'sin(theta)'
y = 'cos(theta)'
print("{} and {} are the best".format(x, y))
```

```
# f-string , usable from 3.6 v of python
x = 10
y = 'books'
z = 'stormlight archives'

print(f'there are {x} {y} of {z} to be released')
```
