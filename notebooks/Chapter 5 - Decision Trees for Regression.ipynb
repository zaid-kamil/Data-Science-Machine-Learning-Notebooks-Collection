```
import pandas as pd
```

```
df = pd.read_csv('/content/drive/My Drive/courses/data analytics basics/datasets/road_accident.csv')
```

```
df
```

```
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```
X = df.iloc[:,:-1]
y = df['FATALSUM']

trainX, testX, trainY, testY = train_test_split(X,y,test_size=.3,random_state=0)
```

```
for i in range(5,25):
    model = DecisionTreeRegressor(max_depth=i,)
    model.fit(trainX, trainY)
    print(i, model.score(testX,testY) * 100)
```

```
model = DecisionTreeRegressor()
model.fit(X, y)
```

```
ypred = model.predict(X)
```

```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
f,ax = plt.subplots(figsize=(15,5))
sns.distplot(y,hist=False,ax=ax,color='red')
sns.distplot(ypred,hist=False,ax=ax)
plt.show()
```
