# Dictionary
- ordered and unindexed datatype
- every value is stored with a name as `Key : value` pairs
- keys are unique and immutable
- values are of any data type and can be same
- python has a function called `dict( )`
- python uses curly braces `{ }` for dictionary

# traversing a dictionary

```python
x
```

```python
# only keys
for i in x:
    print(i)
```

```python
# value can be accessed using
for i in x:
    print(x[i])
```

```python
# key and value both
for k in x:
    print(k , x[k])
```

```python
# best way to loop through dictionary
for k,v in x.items():
    print(k, v)
```

```python
for name,val in y.items():
    print(f"{name.ljust(10)} => {val}")
```
