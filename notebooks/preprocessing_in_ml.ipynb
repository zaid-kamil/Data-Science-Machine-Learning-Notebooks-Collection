# the concept of sklearn preprocessors

1. loading the data
2. sklearn preprocessing package
    - Convert string values to Numeric values
        - Label Encoder 
        - One Hot Encoder
    - Handle missing values
        - Imputer
    - Scale the column (feature scaling)
        - Standard Scaler 
 3. Working with sklearn is different
    - first u get the data
    - the you create an object of a particular class of preprocessor using constructor
    ```
    from sklearn.preprocessing import LabelEncoder
    species_enc = LabelEncoder()
    ```
    - the you fit the data, meaning make preprocessor understand and calculate the data of a column(s)
    ```
    species_enc.fit(dataframe['species'])
    ```
    - then u get the converted data from the preprocessor using the transform functions
    ```
    dataframe['species'] = species_enc.transform() 
    ```
    or use shorter version 
    ```
    dataframe['species'] = species_enc.fit_transform(dataframe['species']
    ```

# Lets do some code

````
import pandas as pd
import numpy as np
import seaborn as sns
````

````
df = sns.load_dataset('penguins')
````

````
df.info()
````

## handle missing values

````
from sklearn.impute import SimpleImputer
````

````
imputer = SimpleImputer()
````

````
cols = ['culmen_length_mm','culmen_depth_mm','flipper_length_mm','body_mass_g']
df[cols] = imputer.fit_transform(df[cols])
````

````
df['sex'].replace(np.nan,df['sex'].value_counts().idxmax(), inplace=True)
````

````
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
````

````
gender_label_encoder = LabelEncoder()
df['sex'] = gender_label_encoder.fit_transform(df['sex'])
````

````
df.head()
````

````
species_hot_encoder = OneHotEncoder()
species_cols = species_hot_encoder.fit_transform(df[['species']]).toarray()
````

````
island_hot_encoder = OneHotEncoder()
island_cols = island_hot_encoder.fit_transform(df[['island']]).toarray()
````

````
species_cols
````

````
island_cols
````

````
df
````

Join the data and dummies

````
data = np.append(species_cols,island_cols,axis=1)
````

````
data = np.append(data ,df[['culmen_length_mm','culmen_depth_mm','flipper_length_mm','body_mass_g','sex']],axis=1)
````

Scaling of data

````
from sklearn.preprocessing import StandardScaler
````

````
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
scaled_data
````
