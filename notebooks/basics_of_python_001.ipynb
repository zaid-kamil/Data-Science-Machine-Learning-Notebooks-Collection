### the 5 ways to print formatted output in python
1. using comma system
2. concatenation of content
3. format specifier using format string
4. using format() method
5. using f-strings âœ“

```
name= "Hugh"
age = 43
surname = 'Jackman'
```

```
print('my name is',name,surname,'and i am',age,'year old')
```

```
print('My name is %s %s and i m %d year old'%(name,surname,age))
```

```
print('my name is '+name+" "+surname+' and i am '+str(age)+' year old') # the difficult one
```

```
print('my name is {} {} and i am {}'.format(name, surname, age)) # until 3.5 python version
```

```
print('my name is {name} {surname} and i am {age}') # nothing happens because python needs to know where are variables
```

```
print(f'my name is {name} {surname} and i am {age}') # whatever is in braces is treated as variables
```

#### wap to calculate the bill at the restaurant

```
item1  = input('enter 1st item : ')
price1 = int(input('enter 1st item price : '))
item2  = input('enter 2nd item : ')
price2 = int(input('enter 2nd item price : '))
item3  = input('enter 3rd item : ')
price3 = int(input('enter 3rd item price : '))
print("Bill :")
print(f'{item1}:{price1}')
print(f'{item2}:{price2}')
print(f'{item3}:{price3}')
total = price1 + price2 + price3
print(f'total cost is {total}')
print(f'GST tax : 30%')
print(f'Amount to be paid : {total + total *.3}')
```
