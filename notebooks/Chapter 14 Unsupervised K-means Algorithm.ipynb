```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
plt.style.use('ggplot')
```

```
df = pd.read_csv('/content/drive/My Drive/courses/data analytics basics/datasets/Mall_Customers.csv')
```

```
df
```

```
df.plot.scatter(x='Annual Income (k$)',y='Spending Score (1-100)')
plt.show()
```

```
from sklearn.cluster import KMeans
```

```
X = df[['Annual Income (k$)','Spending Score (1-100)']]
```

```
wcss = []
for k in range(1,11):
    model = KMeans(n_clusters=k)
    model.fit(X)
    wcss.append(model.inertia_)
```

```
plt.plot(range(1,11),wcss,'ro--')
plt.title("WCSS analysis")
```

```
model = KMeans(n_clusters=5)
model.fit(X)
df['groups']=model.predict(X)
```

```
df.plot.scatter(x='Annual Income (k$)',
                y='Spending Score (1-100)',
                c='groups',
                cmap='rainbow',
                s=100,
                figsize=(13,8))
plt.show()
```

## Age vs Spending Score

```
X = df[['Age','Spending Score (1-100)']]
wcss = []
for k in range(1,11):
    model = KMeans(n_clusters=k)
    model.fit(X)
    wcss.append(model.inertia_)
plt.plot(range(1,11),wcss,'ro--')
plt.title("WCSS analysis")
```

```
model = KMeans(n_clusters=4)
model.fit(X)
df['agewisegroups']=model.predict(X)
```

```
df.plot.scatter(x='Age',
                y='Spending Score (1-100)',
                c='agewisegroups',
                cmap='rainbow',
                s=100,
                figsize=(13,8))
plt.show()
```
