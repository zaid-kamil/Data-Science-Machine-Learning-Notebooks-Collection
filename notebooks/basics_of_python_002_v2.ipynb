```python
for item in range(5):
    print(item)
```

1. range(size)
    - generate a sequence of values from 0 to size-1
2. range(start,stop)
    - generate a numerical sequence from start to stop-1
3. range(start, stop, step)
    - generate a numerical sequence from start to stop -1 but with gap of step size
    

```python
for i in range(5,10):
    print(i)
```

```python
for i in range(5,15,3):
    print(i)
```

```python
# reverse
for i in range(10,0,-1):
    print(i,end=' ')
```

### Break and continue

```python
students = ['alexa','alex','','ajay','alice','','baldev','baloo','','cynthia','ceril']
for name in students:
    if not name:
        print("blank entry found, list ko shi kro")
        break
    print(name)
```

```python
students = ['alexa','alex','','ajay','alice','','baldev','baloo','','cynthia','ceril']
for name in students:
    if not name:
        continue
    print(name)
```

```python
students = ['alexa','alex','','ajay','alice','','baldev','baloo','','cynthia','ceril']
for name in students:
    if name:
        print(name)
```

### for - else and while - else
special loop variation which allows us to check whether loop has successfully iteration all iteration or not 

```python
for i in range(5):
    print(i,end=' ')
else:
    print("completed")
```

```python
for i in range(10):
    print(i,end=' ')
    if i == 3:
        break
else:
    print('completed')
```

```python
# prime number
num = int(input('enter a number to check if its prime or not =>'))
for i in range(2,num):
    print(f"dividing {num}/{i} = {num % i}")
    if num % i == 0:
        print(f'{num} not prime')
        break
else:
    print(f"{num} is PRIME")
```

```python
i = 6
while i > 0:
    if i == 13:
        break
    print(i)
    i-=1
else:
    print("while loop completed")
```

for loop special functions
- enumerate
    - gives us both index and value of item in the loop
- zip
    - used when we want to iterate more than 1 iterators in a single loop

```python
x = [13,15,12,11,12,122,33]
for i in x:
    print(i)
```

```python
for idx,val in enumerate(x):
    print(idx, val)
```

```python
x = ['red','blue','green','yellow','purple','pink','brown','black','white','orange','cyan','magenta']
for i,v in enumerate(x):
    print(v,end=' ')
    if (i+1) % 3 == 0:
        print()
```

```python
colors = ['red','blue','green','yellow','purple','pink','brown','black','white','orange','cyan','magenta']
codes = ['r','b','g','y','p','pk','br','bl','w','o','c','m']
```

```python
for color,code in zip(colors,codes):
    print(f'{code} stands for {color}')
```

```python
x = [1,2,3,4,5,6,76,7,8,8]
y = [12,2,321,3,12,31,23,123,12]
z = [2,3,3,1,3,2,2,2,2,2,3,3]

for i,j,k in zip(x,y,z):
    print(i+j+k)
```
