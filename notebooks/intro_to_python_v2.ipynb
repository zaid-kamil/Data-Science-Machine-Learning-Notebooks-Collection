```
person = 'bruce wayne'
secret_identity = 'Batman'
```

```
print(person,'is',secret_identity) # type 1 using comma seperation
```

```
print("%s is %s" %(person,secret_identity)) # type 3 using format specifier ( c based concept)
```

```
print("{} is {}".format(person, secret_identity)) # type 4 using format fuction of string (old method)
```

```
print(f'{person} is {secret_identity}') # type 5 using f strings available from python 3.6 
```
