subset using conditions

```
pokemon['Type 1'] == 'Bug'
```

```
pokemon[pokemon['Type 1'] == 'Bug']
```

```
condition = pokemon['Type 1'] == 'Electric'
pokemon[condition] # only electric
```

```
pokemon[~condition] # dataset except electric
```

```
condition = pokemon['Type 1'].isin(['Dragon','Fairy'])
pokemon[condition]
```

```
islessHP =  pokemon['HP'] < 50
fig = px.scatter(data_frame=pokemon[islessHP], x='Attack',y='Defense',color='Stage',title='pokemons with less than 50 HP', hover_name='Name')
fig.show()
```

```
islessHP =  pokemon['HP'] < 50
fig = px.scatter_3d(data_frame=pokemon[islessHP], x='Attack',y='Defense',z='Speed',color='Stage',title='pokemons with less than 50 HP', hover_name='Name')
fig.show()
```
