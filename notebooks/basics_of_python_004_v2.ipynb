# Dictionaries
- unindexed
- ordered
- heterogenous
- dynamic and mutable
- data is store in `KEY`:`VALUE` format
- `KEY` is always unique and immutable
- `VALUE` can be anything except `set` type
- represented by `{ }`
- constructor is `dict( )`

## Objective
- creation
- methods

```
marks = {
    'shivansh':20,
    'mahi':19,
    'archit':19,
    'saksham':19,
    'akshita':17,
    'swati':16
}
```

```
emp = {'name':'rahul','age':27,'quote':'naam to suna hoga', 'married' : False}
print(emp)
```

```
shop = dict(name = 'ICECREAM SHOP',address='hazratganj, lucknow',owner = 'rahul')
print(shop)
```

```
print(marks)
```

#### accessing the data from dict

```
 shop['address']
```

```
shop['type'] #error => keyerror
```

```
if 'type' in shop: # if key exists in shop dict then show
    print(shop['type'])
```

```
if 'owner' in shop: # if key exists in shop dict then show
    print(shop['owner'])
```

```
emp.get('age')
```

```
emp.get('gender') # no error when key is not found
```

```
emp.get('gender','not found') # default value is shown when key is not found
```

#### looping the data from dict

```
# all the keys of dict by default
for i in marks:
    print(i)
```

```
for key in marks:
    print(marks[key])
```

```
for key in marks:
    print(key, marks[key])
```

```
# better loop in dict
for k, v in marks.items():
    print(k,v)
```

## methods
- add key value to dict
- update key value 
- remove key value
- helper functions

```
contacts = {} # empty
```

```
# add
contacts['sam'] = 8392829292
contacts
```

```
contacts['alex'] = 8592829292
contacts
```

```
contacts['samuel'] = [8392829292,8990200221]
contacts
```

```
# update
contacts['sam'] = 8999923939 # key exists but values is new => its updated in dict
contacts
```

```
# update a dict with multiple k:v pairs
contacts2 = {'ravi':9099090902,'harshita':9383832221}
contacts.update(contacts2)
contacts
```

```
# remove value by keys
if 'sam' in contacts:
    contacts.pop('sam') # pop works as remove in dict
contacts
```

```
contacts.popitem() # removes last item from dict
```

```
print(contacts)
```
