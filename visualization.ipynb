### Bar plot

```python
years = list(range(2001, 2022))
rainfall = [1000,1200,1004,1050,1020,1100,1120,
            1060,1220,1050,1050,1020,1120,1220,
            1050,1230,1204,1150,1120,1110,1220,]
len(years) == len(rainfall)
```

```python
plt.bar(x=years,height=rainfall,width=.6,color='red')
plt.show()
```

```python
plt.bar(x=years,height=rainfall,color='red',alpha=.6)
plt.show()
```

```python
fig, ax = plt.subplots(ncols=2,figsize=(12,4))
ax[0].plot(years, rainfall, 'rD:')
ax[1].bar(years, rainfall, color='lightgreen')
plt.show()
```

```python
plt.barh(years, rainfall, color='lightgreen')
plt.show()
```

```python
names = ['Mahendra Singh Dhoni','Muthaiya Murlidharan','Mahela Jayavardhana']
scores = [2193,900,1500]
plt.barh(names,scores,color=['red','green','blue'])
plt.show()
```

```python
fig ,ax = plt.subplots()
rects = ax.barh(names,scores,color=['red','green','blue'])
ax.bar_label(rects,padding=3)
plt.xlim(right=2500)
plt.show()
```

```python
years = list(range(2010, 2022))
rainfall = [1000,1200,1004,1050,1020,1100,1120,
            1060,1220,1050,900,800]
len(years) , len(rainfall)
```

```python
fig,ax = plt.subplots(figsize=(15,5))
rects = ax.bar(years, rainfall,color='skyblue',hatch='o',edgecolor='red')
ax.bar_label(rects, padding=3, fontsize=10)
ax.set_ylim(top=1350)
plt.show()
```

```python
countries = ['China','India','Us','Indonesia','Pakistan']
pop = [1439323776,1380004385,331002651,273523615,220892340]
```

```python
plt.pie(pop, labels=countries,radius=2,startangle=90)
plt.show()
```

```python
plt.pie(pop, labels=countries,radius=2,startangle=90, counterclock=False)
plt.show()
```

```python
explode = [0, 0, 0.2, 0.2, 0.2]
plt.pie(pop, labels=countries,radius=2,startangle=90, counterclock=False,explode=explode)
plt.show()
```

```python
explode = [0, 0, 0.2, 0.2, 0.2]
plt.pie(pop, labels=countries,radius=2,startangle=90, 
        counterclock=False,explode=explode, autopct='%.1f%%')
plt.show()
```

```python
plt.figure(facecolor='#89f9d8')
explode = [0, 0, 0.2, 0.2, 0.2]
plt.pie(pop, labels=countries,radius=2,startangle=90, 
        counterclock=False,explode=explode, autopct='%.1f%%',
        textprops={'fontsize':15,'color':'white'})
plt.show()
```

```python
explode = [0, 0, 0.2, 0.2, 0.2]
plt.pie(pop, labels=countries,radius=2,startangle=90, 
        counterclock=False,explode=explode, autopct='%.1f%%',
        textprops={'fontsize':15}, wedgeprops={'width':1,'edgecolor':'white','hatch':'/'},
        pctdistance=.7)
plt.show()
```

```python

plt.pie(pop, labels=countries,radius=2 ,
        autopct='%.1f%%',textprops={'fontsize':15},
        wedgeprops={'width':1,'edgecolor':'white','hatch':'/'},
        pctdistance=.7)
plt.pie([23,20,25],labels=['A','B','C'], labeldistance=.5)
plt.show()
```
