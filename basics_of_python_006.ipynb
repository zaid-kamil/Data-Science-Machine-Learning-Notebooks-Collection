# File handling

- `open()` builtin function
- `os` module usage

### reading a file

```
with open('/content/sample_data/README.md') as file:
    print(file.read())
```

### write data to a file

```
newfile = open('file1.txt','w')
newfile.write('The best movies are')
newfile.write(('to be continued...'))
newfile.close() # save and close
```

```
# file will autosave and close after block ends
with open('file1.txt','w') as newfile:
    newfile.write('The best movies are\n')
    newfile.write(('Captain America : Winter Soldier\n'))
    newfile.write(('Avengers Endgame\n'))

# w mode overwrites the file
```

### updating a file with new data

```
with open('file1.txt','a') as file:
    file.write('Iron Man\n')
    file.write('Sucide Squad\n')
    file.write('Interstellar\n')
```

### write and read or vice versa in the same session

```
file =  open('file1.txt','w+')
file.write('Iron Man\n')
file.write('Sucide Squad\n')
file.write('Interstellar\n')
file.seek(0) # takes the cursor to start of file
print(file.read())
file.close()
```
