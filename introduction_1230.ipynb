# Output in python
- raw output
    - when we directly put a variable in the code cell, and execute it, the values of that variable are shown as a raw value
- formatted output
    - when we want to change the output and make it readable for users
    - we use `print()` function for formatted output
    - we have different technique in `print()` function for making output better
    - in python 3.6 a special type of string concept was added called `f-string`, that make displaying string and variable easy

````python
name = "Alexa"
birthday = "3 march"
age = 10
````

**style 1 - comma seperated value**

````python
print(name, birthday, age)
print(a,b,c,d,city,num)
print("You are", name, "and your age is", age) # easiest
````

**style 2 - use c like format specifier => %s | %d | %f | %.2f**

````python
print("%s %s %d"%(name,birthday,age))
print('You are %s and %s, is your birthday and you will be %d yrs old'%(name,birthday,age))
````

**style 3 - use java like concatenation using + operator => not the good way**

````python
print('You are '+name+' and '+birthday+', is your birthday and you will be '+str(age)+' yrs old')
````

**style 4 - using format function of the string => (old)**

````python
print('You are {} and {}, is your bday and you will be {} yrs old 😊'.format(name,birthday,age))
````

**style 5 - using f-string => best way**

````python
print(f'You are {name} and {birthday}, is your bday and you will be {age} yrs old 👌')
````

````python
print(f'{a} + {b} = {a + b}')
````

````python
print(f'{a} divided {b} = {a / b}')
````
