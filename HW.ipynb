**SIMPLE LINEAR REGRESSION**

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as pt
import plotly.express as px

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```
df = pd.read_csv('/content/Salary_Data.csv')
df.head()
```

```
df.info()
```

```
df.describe()
```

```
sns.distplot(df['YearsExperience'])
plt.show()
```

```
sns.distplot(df['Salary'])
plt.show()
```

```
fig = px.scatter(df,x='YearsExperience',y='Salary',size='Salary',width=800)
fig.show()
```

SPLITTING

```
X = df[['YearsExperience']]
X.shape
```

```
y = df[['Salary']]
y.shape
```

```
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=.2,random_state=0 )
```

```
print("data distribution : \nX train",xtrain.shape,"\nY train",ytrain.shape,"\nX test",xtest.shape,"\nY test",ytest.shape)
```

**FITTING THE DATASET**

```
reg = LinearRegression()
```

```
reg.fit(xtrain, ytrain)
```

**EVALUATION OF MODEL**

```
reg.score(xtest, ytest) * 100
```

```
ypred = reg.predict(xtrain)
```

```
plt.plot(xtrain, ypred,label='prediction',color='red')
plt.scatter(xtrain, ytrain,label='original')
plt.title("original vs prediction on training data")
plt.legend()
plt.show()
```

```
ypred= reg.predict(xtest)
plt.plot(xtest, ypred,label='prediction',color='red')
plt.scatter(xtest, ytest,label='original')
plt.title("original vs prediction on training data")
plt.legend()
plt.show()
```

TRAINING OF DATASET WRT ALGORITHM

```
model = LinearRegression()
model.fit(X,y)
# last 2 lines not neccessary
ypred = model.predict(X)
df['predicted'] = ypred
```

```
ax = df.plot.scatter(x='YearsExperience',y='Salary',c='red',label='prediction')
df.plot.line(x='YearsExperience',y='predicted',label='original',ax=ax)
plt.show()
```

ROOT MEAN SQUARE ERROR

```
RMSE = np.sqrt(mean_squared_error(df['Salary'], df['predicted']))
```
