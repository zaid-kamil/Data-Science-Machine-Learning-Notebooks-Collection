validation function

```python
val = input("enter something: ")
if val.isalpha():
    print("you entered only alphabets")
if val.isnumeric():
    print('you entered only numbers')
if val.isalnum():
    print('you entered digits and alphabets')
if val.isspace():
    print("your string contains only spaces")
if val.isupper():
    print('string is in uppercase')
if val.islower():
    print('string is in lowercase')
```

examples

wap to remove all the vowels in a string

```python
data = input("enter something: ")
for vowel in "aeiouAEIOU":
    data = data.replace(vowel,'')

print(data)
```

wap to replace all the vowels in a string with a _

```python
data = input('enter a message')
for v in "aeiouAEIOU":
    data = data.replace(v,'_')
print(data)
```

wap to sort a list of guest and remove names without honorifics

```python
guestlist = ['Mr. Raju','Dr. Verma','Ajay Singh','Mrs Verma','Vijay']

for guest in guestlist:
    if guest.startswith("Mr") or guest.startswith('Mrs') or guest.startswith("Dr"):
        print(f'welcome, {guest}')
    else:
        print(f'throw {guest} out!')
```

wap to count all the vowels in every word of a sentence

```python
sentence =input("enter a lengthy msg=>")
words = sentence.split()
for word in words:
    vowel_count = 0
    for v in 'aeiouAEIOU':
        vowel_count+= word.count(v)
    print(f'{word.ljust(15)} {vowel_count}')
```

wap to find all the indexes of search string

```python
msg = 'this is an encylopedia which is not wikipedia.this is an encylopedia which is not wikipedia.'
search = input("search anything: ")
start = 0
for i in range(msg.count(search)):
    idx = msg.find(search,start)
    print(f'{search} found at {idx} index')
    start =  idx+1
```
