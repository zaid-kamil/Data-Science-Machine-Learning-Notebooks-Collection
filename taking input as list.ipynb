# Lets combine all of this in one line

```
numbers= list(map(int,input('numbers>>').split(',')))
```

```
print(type(numbers),numbers)
```

or

**set**

```
numbers= set(map(int,input('numbers>>').split(',')))
numbers
```

**tuple**

```
numbers= tuple(map(int,input('numbers>>').split(',')))
numbers
```
