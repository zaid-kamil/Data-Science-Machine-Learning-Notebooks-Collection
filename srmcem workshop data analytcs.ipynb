```python
string = "hello"
reversed_string = ""
for char in reversed(string):
  reversed_string+=char
print(reversed_string)
```

```python
def reverse_string(string):
  reversed_string=""

  for char in reversed(string):
    reversed_string += char

  return reversed_string
string = "hello"
reversed_string = reverse_string(string)
print(reversed_string)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
x=[1,2,3,4,5]
y=[10,14,18,24,30]
fig = px.line(x=x, y=y, title='Line Chart', labels={'x':'X',"y":'Y'})
fig.show()
```

```python
samples = ['A','B','C','D','E']
counts = [10,14,18,24,30]
fig = px.bar(x=samples, y=counts, title = 'Bar Chart' , labels={'x':'Sample','y':'Counts'}, color=['A','B','C','D','E'],
            height=400, width=500 )
fig.show()
```

```python
labels = ['category A','category B','category C','category D','category E']
values = [10,14,18,24,30]
fig = px.pie(names=labels,values = values, title = 'Pie Chart', labels={'names':'Category', 'values':'Count'},hole=0.3,
             height=400, width=500,template='plotly_dark')
fig.show()
```
