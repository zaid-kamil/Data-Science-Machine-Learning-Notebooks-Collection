```
data = ['apple','banana','cherry','apricot','grapes']
```

```
### wap to only list fruits containing letter 'e'

for fruit in data:
    if 'e' in fruit:
        print(fruit)
```

```
# wap to create a list of fruits starting with 'a' letter
```

```
data_2 = []
for fruit in data:
    if fruit.startswith('a'):
        data_2.append(fruit)
print(data_2)
```
