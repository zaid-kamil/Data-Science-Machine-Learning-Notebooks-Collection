# functions - we have 11 methods in list datatype
- adding items
    - append() - add single item to end of list
    - insert() - add single element to specific index of list
    - extend() - add a list to an existing list
- removing items
    - remove() - remove single occurance of a value in list
    - pop()    - remove value on a particular index
    - clear()  - clears all items in a list
- helper functions
    - count()  -  count element occurance in a list
    - index()  - gets the index of element search in list
    - sort()   - sorts a homogenous list in ascending/descending order
    - reverse()- reverses a list
    - copy()   - copies a list

# comprehension
its the way to create a new list from an existing list with one line of code logic

````python
# wap to add 10 to each element of a list and store result in new list
x = [1,2,3,4,5,6,7,8,9]
x2 = [] 
for i in x:
    j = i + 10
    x2.append(j)
print(x)
print(x2)
````

````python
# comprehension
y = [23,54,12,45,213,12,2,23,5]
y10 = [i + 10 for i in y]
print(y)
print(y10)
````

````python
# wap to create a list of squares from existing list
a = [2,3,4,5,6,]
a2 = [i**2 for i in a]
print(a)
print(a2)
````

````python
# wap to filter a list of item so that any value lower than 50 is removed
x = [102,23,320,122,120,53,20,10,30]
x5 = [i for i in x if i > 50]
print(x)
print(x5)
````

# map and filter with lambda

````python
a = [12,43,2,10,20,13]
````

````python
map(lambda i: i**2, a) # lazy object
````

````python
list(map(lambda i: i**2, a))
````

````python
tuple(map(lambda i: i**2, a))
````

````python
set(map(lambda i: i**2, a))
````
