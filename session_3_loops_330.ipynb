Two types of loops are available to us in python
- for loop
    - this loop is used when the number of iterations are knows. also called **step based** loop
- while loop
    - this is used when we want to check a condition and then perform some task in a loop. the number of steps are unknown and depends on the condition. Also called **condition based loop**

- this sequence can be used for the forloop or to make a list/tuple/set data
- **it has three styles** of usage
    - range(stop_value)

break and continue

```python
name = "Vijay deenanath chauhan"
for char in name:
    if char == 'c':
        break # stop the loop
    print(char)
```

```python
for char in name:
    if char == 'c':
        continue #skip to next item 
    print(char)
```

```python
products = ['potato','tomato','rotten potato','carrot','rotten tomato','broken eggs']
for item in products:
    if 'rotten' in item:
        continue
    print(f'we got {item}')
```

for - else loop
- this is used to confirm if the loop is completed or not. the code in else part is only executed when the loop completes all the iterations otherwise if the loop breaks before completion, the else part will be skipped

```python
# enter five animal names
for i in range(5):
    name = input('>>')
    if len(name) == 0:
        print('very bad! you get 0 points')
        break
else:
    print('you get 10 points')
```

```python
# prime number check - those numbers that are not divisible by other numbers between 1 and itself
num = 60
for i in range(2,num):
    # print(f'{num}/{i} = {num%i}')
    if num % i == 0:
        print('not a prime')
        break
else:
    print('its prime')
```

# while loop
when u want to repeat something based on a condition

```python
a = 1
while a < 10:
    print(f'hello {a}')
    a+=1
```

```python
# sum of digits
number = int(input('enter a number:'))
total = 0
while number > 0:
    r = number % 10
    total += r
    number //= 10
print(total)
```

```python
# infinite loops with break
while True:
    name =  input('who are you? ')
    if name == 'raju':
        print('welcome master')
        break
    else:
        print('you are not allowed!')
```

```python
# sum of all values entered by user
donations = 0
while True:
    name = input('Your name= ')
    amt = int(input('Your dontation = '))
    if amt>0:
        print(f'{name} donated {amt} rupees ðŸ¤©')    
        donations += amt
    if not name:
        break
print(f'total donations: {donations}')
```

```python
# multiline input
content = ""
while True:
    line = input('>> ')
    if line:
        content += line +'\n'
    else:
        break
print(content)
```
