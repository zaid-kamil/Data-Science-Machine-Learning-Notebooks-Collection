### utility functions in set

```
civil_war = {'captain america',
             'winter soldier',
             'black panther',
             'black widow',
             'iron man'}
avengers = {'captain america',
             'winter soldier',
             'black panther',
             'black widow',
             'iron man',
             'hulk',
             'thor',
             'dr strange'}
```

```
civil_war.union(avengers)
```

```
civil_war.intersection(avengers) # common to both
```

```
avengers.difference(civil_war)
```
