```
mobiles ={
    'vivo y71':12000,
    'one plus 7t':38000,
    'oppo f9':18000,
}
```

```
for name, price in mobiles.items():
    print(name, price)
```

```
# total price of all mobiles
tp = 0
for name, price in mobiles.items():
    tp += price
print('total price',tp)
```

```
tp = sum(mobiles.values())
print("total price",tp)
```
