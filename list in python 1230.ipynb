## list functions for utility (helper functions)
- count() => counts the occurance of a value in a list

````python
# wap to print a fibonacci series
fib = [0,1]
for i in range(10):
    fib.append(fib[-1] + fib[-2])
print(fib)
````

````python
# wap to remove all the values which are repeated in list
x = [1,2,3,3,5,5,4,4]
temp = x.copy()
for item in temp:
    counter = x.count(item)
    if counter > 1:
        for i in range(counter):
            x.remove(item)
print(x)
````

more examples of map and filter functions

````python
names = ['Alex','Alan','Oliver',"Eliza","Martha",'Sam']
````

````python
# normal version
names_with_a = []
for name in names:
    if "a" in name:
        names_with_a.append(name)
names_with_a
````

````python
# one liner
names_with_a = filter(lambda nm : "a" in nm, names)
list(names_with_a)
````

````python
# normal
name_sizes = []
for nm in names:
    size = len(nm)
    name_sizes.append(size)
name_sizes
````

````python
# one liner
name_lenghts = map(lambda nm: len(nm), names)
list(name_lenghts)
````
