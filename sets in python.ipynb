# Sets 
- unordered data structure 
- unindexed
- every value is unique
- mutable - we can add/remove data
- represented with `{  }`
- empty set is created with `set()`
- `set()` function is used to convert other dataset in set datatype

## set functions
- add data
    - add
    - update
- remove data
    - remove
    - discard
    - clear
- utility 
    - union
    - intersection
    - difference
    - symmetric_difference
- validation
    - issubset
    - issuperset
    - isdisjoint


```
# single value entry
names= {'Alex','Alice'}
names.add('Andrew') 
names.add('Amber')
names.add('Ankita')
names.add('Amar')
names.add('Alex')
names.add('alice') # case sensitive, thus added
names
```

```
# multi valued entry
names.update({'Bison','Bulma'}) # add a set to an existing set
```

```
somemorenames = ['Bikram','Bikrant','Bishvajeet','batman']
names.update(somemorenames) # add a list to existing set
print(names)
```

```
 names.remove('Alex') # removes value from set if found else error
```

```
names.discard('Bikrant') # removes  value from set if found else nothing # better
```

```
print(names)
```

```
names.clear() # remove all values
print(names)
```
