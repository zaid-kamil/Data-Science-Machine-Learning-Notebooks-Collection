```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = sns.load_dataset('penguins')
data.head()
```

```python
data.info()
```

```python
sns.countplot(x='species',data=data)
plt.show()
```

```python
sns.countplot(x='species',data=data, hue="island")
plt.show()
```

```python
plt.scatter(x='bill_length_mm',
            y='bill_depth_mm',
            data= data)
plt.show()
```

```python
px.scatter_3d(data,
              x='bill_length_mm',
              y='bill_depth_mm',
              z='flipper_length_mm',
              color='species')
```
