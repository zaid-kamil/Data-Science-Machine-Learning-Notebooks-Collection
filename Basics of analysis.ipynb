```python
dfc.head(2)
```

removing columns

```python
cols_to_remove = ['Type','Coverage','AREA','REG','DEV']
dfc.drop(columns=cols_to_remove,axis=1, inplace=True)
```

```python
dfc.head(3)
```

```python
dfc.shape
```

renaming columns

```python
dfc.rename({
    'OdName' :'Country',
    'AreaName':'Continent',
    'RegName': 'Region',
    'DevName':'CountryStatus'
},axis=1, inplace=True)
```

adding a new column

```python
dfc['Faltu'] = "No Idea"
```

```python
dfc.head(2)
```

```python
# actual impletation
cols_for_sum = range(1980,2014)
dfc['total']=dfc[cols_for_sum].sum(axis=1)
```

```python
dfc.head(2)
```

```python
dfc.drop(columns=['Faltu'],axis=1,inplace=True)
```

changing the index

```python
dfc.set_index('Country',inplace=True)
```

```python
dfc['total'].plot(kind='line')
plt.xticks(rotation=90)
plt.show()
```

updating values using function and apply

```python
dfc['CountryStatus'].value_counts()
```

```python
dfc['CountryStatus'].value_counts().plot(kind='pie',autopct='%.1f%%')
plt.show()
```

```python
def update_col_val(value):
    if 'Developing' in value:
        return 0
    elif 'Developed' in value:
        return 1
```

```python
dfc['CountryStatus'] = dfc['CountryStatus'].apply(update_col_val)
```

```python
dfc.head(3)
```

sorting

```python
dfc.sort_values(by='total',ascending=False,inplace=True)
```

```python
dfc.head()
```

```python
dfc['total'].head(10).plot(kind='barh',title='Immgration 1980-2013')
plt.show()
```
