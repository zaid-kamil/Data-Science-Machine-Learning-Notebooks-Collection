# Tuple - datatype
- characteristics
    - [x] indexed -> every element has a index number
    - [x] ordered -> sequence is maintained
    - [x] **immutable** -> we cannot add or remove values
    - [x] heterogenous - > data can be mixed type
    - [x] can have duplicate values
    - [x] has a function/constructor called `tuple`
    - [x] represented by `( 1,2,3 )` round brackets



# what we are going to learn
- creation
- indexing and slicing
- functions
- looping

# Creation

```python
x = (1,2,3,4) # tuple
y = 2,3,4,5   # tuple
```

```python
print(x)
```

```python
print(y)
```

```python
a = ((1,2,3),(3,3,4),(1,23,2))
a
```

```python
word = 'Hippopotamus'
```

```python
chars = tuple(word)
chars
```

```python
x = [1,2,3,4]
xt = tuple(x)
xt
```

# indexing & slicing

```python
chars[:5]
```

```python
chars[2:-2]
```

# looping

```python
for i in a:
    print(i)
```

```python
for i in x:
    print(i)
```

```python
for i in x[::-1]:
    print(i)
```
