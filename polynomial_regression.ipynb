# polynomial regression

```
import pandas as pd
import numpy as np
```

```
import matplotlib.pyplot as plt
```

```
#dataset name
filename = '/content/drive/My Drive/courses/data analytics/datasets/Position_Salaries.csv'
```

```
dataset = pd.read_csv(filename)
dataset.info()
dataset.head()
```

```
dataset.plot(kind='line',x='Level',y='Salary',style='o',xlim=(0,11))
```

```
X = dataset[['Level']]
y = dataset['Salary']
```

```
from sklearn.linear_model import LinearRegression
```

```
reg = LinearRegression()
reg.fit(X,y)
```

```
plt.plot(X,y,'ro')
plt.plot(X,reg.predict(X))
plt.show()
```
