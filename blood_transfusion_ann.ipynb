```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df =  pd.read_csv('https://raw.githubusercontent.com/digipodium/Datasets/main/classfication/blood_transfusion.csv')
```

```python
df.head()
```

```python
X = df.drop(columns='Class')
y = df.Class
```

```python
targetEnc = LabelEncoder()
y = targetEnc.fit_transform(y)
```

```python
X.shape ,y.shape
```

Creating a neural network

```python
from tensorflow.keras.models import Sequential # a nn model
from tensorflow.keras.layers import Dense # a layer of neuron
from tensorflow.keras.optimizers import SGD # stochastic gradient descent
```

```python
model = Sequential()
```

```python
model.add(Dense(15,activation='relu', input_shape=(4,)))
```

```python
model.add(Dense(5, activation='relu'))
```

```python
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer=SGD(), loss='binary_crossentropy', metrics=['accuracy'], )
```

training the model

```python
model.fit(x=X, y=y, batch_size=10, epochs=10)
```

```python
model.save('blood_transfusion.h5')
```
