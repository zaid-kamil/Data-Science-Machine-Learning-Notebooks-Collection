```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
df = pd.read_csv('/content/drive/My Drive/courses/data analytics basics/datasets/Mall_Customers.csv')
df.head()
```

```
df.columns = ['id','genre','age','income','spend_score']
df
```

```
X = df[['income','spend_score']]
```

# dendogram

```
plt.style.use('seaborn')
```

```
import scipy.cluster.hierarchy as sch
```

```
plt.figure(figsize=(12,8))
dendogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.grid(True)
plt.show()
```

```
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering(n_clusters = 5)
```

```
y_pred = model.fit_predict(X)
df['groups'] = y_pred
df
```

```
df.plot.scatter(x='income',y='spend_score',c='groups',cmap='rainbow')
```
