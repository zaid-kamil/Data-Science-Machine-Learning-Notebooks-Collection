# Dictionary Functions
retrieve value from dict 
- get()
- values()
- keys()
- items()

adding value to dict
- update()
- **we add values to dict directly**

removing value from dict
- pop()
- popitem()
- clear()

other utility function
- fromkeys()
- setdefault()
- copy()

```python
shopping_list = {
    'flour':'5kg',
    'sugar':'10kg',
    'rice':'10kg',
    'oil': '2pcs',
    'sanitizer':'10 bottles',
    'soaps':'10'
}
```

```python
list(shopping_list.items()) # key value list
# this is used in loop mainly
```

adding values

```python
shopping_list.update({
    'noodles':'3pc',
    'cheese':'1pc'
})

shopping_list
```

adding single key and value in a dict

```python
shopping_list['milk'] = '3 ltr.'
shopping_list
```

```python
shopping_list['water'] = '5 ltr.'
shopping_list
```

update value in a dict

```python
shopping_list['sugar'] = '2 kg'
shopping_list
```

```python
shopping_list['milk'] = '1 ltr'
shopping_list
```

removing stuff

```python
if 'noodles' in shopping_list:   # if item is in dictionary
    shopping_list.pop('noodles') # key name to remove
```
