```
import numpy as np
import matplotlib.pyplot as plt
from apyori import apriori
import pandas as pd
```

```
data = pd.read_csv('/content/drive/My Drive/courses/data analytics/datasets/Market_Basket_Optimisation.csv',header=None)
data
```

```
data.info()
```

```
transactions = []
for i in range(data.shape[0]):
    transactions.append([str(data.values[i,j]) for j in range(data.shape[1])])
```

```
association_rules = apriori(transactions,min_support=0.005, min_confidence=.3, min_lift=3, min_length=2) 
results = list(association_rules)
```
