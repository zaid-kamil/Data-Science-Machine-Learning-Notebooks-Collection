# slicing

```python
x = np.random.randint(1,100,(10,10))
```

```python
x
```

```python
# first row
x[0,:]
```

```python
# first row, first five cols
x[0,:5]
```

```python
# first row, last five cols
x[0,-5:]
```

```python
# last 3 rows and all cols
x[-3:, :]
```

```python
# all row , column 0
x[:, 0]
```

```python
x[:,-1] # last column
```

```python
x[:,:5] # first five columns
```
