```
import pandas as pd
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```
grocery_df = pd.read_csv('/content/drive/My Drive/courses/data analytics basics/datasets/GROCERY_SHOP.csv')
```

```
X =  grocery_df[['FRESH','FROZEN','GROCERY']]
y = grocery_df['HOSTEL_OR_RETAIL']
```

```
scaler = StandardScaler()
scaledX = scaler.fit_transform(X)
trainX,testX, trainY,testY = train_test_split(scaledX,y,test_size=.2, random_state=0)
clf = SVC()
clf.fit(trainX,trainY)
ypred = clf.predict(testX)
confusion_matrix(testY, ypred)
```

```
ypredfull = clf.predict(scaledX)
confusion_matrix(y, ypredfull)
```

```
clf.score(testX,testY) * 100
```

```
grocery_df['predictions'] = ypredfull
```

```
fig = px.scatter_3d(data_frame=grocery_df, x = 'FRESH', y='FROZEN', 
                    z='GROCERY',color='predictions',symbol='HOSTEL_OR_RETAIL',
                    size='GROCERY',)
fig.show()
```
