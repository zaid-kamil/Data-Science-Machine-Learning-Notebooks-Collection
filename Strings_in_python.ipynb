# tasks

- wap  to count all the vowels in a string entered by user
- wap to replace all vowels in a user inputted string by '#' char
- wap to print all words that start with 'a' char in a sentence

#### wap to count all the vowels in a string entered by user

```
data = input("enter your sentence: ")
vowels = "aeiou"
total = 0
for char in vowels:
    count = data.lower().count(char)
    total += count
print(f"you entered : '{data}'")
print(f"it has {total} vowels")
```

#### wap to replace all vowels in a user inputted string by '#' char

```
data = input("enter your sentence: ")
vowels = "aeiou"
for char in vowels :
    data = data.replace(char, '#')
print("vowels removed")
print(data)
```

#### wap to print all words that start with 'a' char in a sentence

```
sentence = input("enter a sentence : ")
words = sentence.split()
for item in words:
    if item.lower().startswith('a'):
        print(item)
```
