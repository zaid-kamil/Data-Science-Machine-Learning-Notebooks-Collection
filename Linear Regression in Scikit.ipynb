```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('/content/drive/My Drive/courses/data analytics/datasets/Salary_Data.csv')
df.info()
df.head()
```

# simple linear regression

```
# understand the data distribution
plt.plot(df['YearsExperience'],df['Salary'],'ro')
```

```
from sklearn.linear_model import LinearRegression
```

```
salaryRegressor = LinearRegression()
```

```
X = df[['YearsExperience']]
y = df['Salary']
salaryRegressor.fit(X,y)
```

```
salaryRegressor.coef_
```

```
salaryRegressor.intercept_
```

```
salaryRegressor.predict([[3]])
```

```
df['predictions'] =salaryRegressor.predict(X)
```

```
df.head()
```

```
plt.plot(df['YearsExperience'],df['Salary'],'ro')
plt.plot(df['YearsExperience'],df['predictions'],'g')
plt.show()
```

```
import pickle
```

```
# saving model
with open('salary_prediction_model.pk','wb') as file:
    pickle.dump(salaryRegressor, file)
```

```
# load model
with open('salary_prediction_model.pk','rb') as file:
    salary_model = pickle.load(file) 
    print('loaded model')
```
