## validation function
function that will give us True and False based on the check

````python
name = input('enter name:')
if name.isupper():
    print('well your name is in uppercase')
elif name.islower():
    print("Okay, your name is in lowercase")
elif name.istitle():
    print('Name looks like it is in title case')
else:
    print("learn how to format the content")
````

````python
val = input("enter anything:")
if val.isalpha():
    print("you have entered only alphabets")
if val.isalnum():
    print("you have entered alphabets or digits or mixed content")
if val.isnumeric():
    print('you have only numbers')
if val.isspace():
    print('you have given only spaces')
if val.isdecimal():
    print('you have given number in decimal format')
````

````python
guest_name = input("Please enter your name:")
if guest_name.lower().startswith("mr."):
    print(f"welcome Mister {guest_name[4:]}") # skipped the first 4 chars
if guest_name.lower().startswith("ms."):
    print(f'welcome Miss {guest_name}')
if guest_name.lower().startswith("dr."):
    print(f'welcome Doctor {guest_name}')
if guest_name.lower().startswith("Sir"):
    print(f'welcome Sir {guest_name}')
````

````python
file = input('enter file name:')
if file.endswith('.txt'):
    print('TEXT FILE')
if file.endswith('.jpg') or file.endswith(".png") or file.endswith('.webp'):
    print("IMAGE FILE")
if file.endswith(".exe"):
    print("EXECUTABLE FILE")
````
