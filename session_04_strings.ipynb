slicing syntax
```
var[start:end+1]
var[:end+1]  -> starts from 0 index
var[start:]  -> ends on last item
var[:]       -> start from 0 and go till end
var[start:end+1:step] -> step mean the gap
var[::step]  -> go from 0 to end with a specific gap
```

````python
name = "Vijay Deenanath Chauhan"
````

````python
name[0:5]
````

````python
name[:5]  #first 5
````

````python
name[-7:] # last 7
````

````python
name[6:-8]
````

````python
# first and last 2 char skip from string
````

````python
name[2:-2]
````

````python
name[::-1] # reverse string
````

````python
name[::2] # character on even indexes
````

````python
# wap to skip all the vowels and display only consonants in a message
for char in "this is sparta":
    if char not in 'aeiouAEIOU':
        print(char, end='')
````

````python
# wap to print clean string skipping all the punctuations from the string
puncs = ''''!@~#$%^&*()_+{}"><?.,'''
msg = 'this(*&!@#(*!@#)) is n!@#)*@#)(!@*ot sp!@#!>?<#>!@#_arta'
for char in msg:
    if char not in puncs:
        print(char,end='')
````

````python
guests = ['Mr. Baloo','Dr. Singh','Mrs. Pandey','Sir Jadeja','Rahul','Vijay','Ms Singh']
````

````python
for guest in guests:
    if guest.lower().startswith('mr'):
        print(f'welcome {guest}')
    elif guest.startswith("Mrs"):
        print(f'welcome {guest}')
    elif guest.startswith('Dr'):
        print(f'welcome {guest}')
    else:
        print(f'{guest}, you are not allowed')
````
