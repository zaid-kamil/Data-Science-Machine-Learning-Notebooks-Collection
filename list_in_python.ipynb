## functions
- add values in list
    - append => add an item to end of list
    - insert => add an item at the given index value
    - extend => add a list to an existing list
- remove value from list
    - pop => remove an item from an index otherwise the end of list
    - remove => remove item that is given by user
    - clear => remove all items from the list


- utility functions
    - copy => create a copy of list values in a new variable
    - sort => sort the list if all values are of same type
    - count => count the occurance of value in the list
    - reverse => reverses all items in a list
    - index => gives us index value of any item that is found

wap to add 15 numbers to a list and then display the sum and avg of that list

```
x = []
for i in range(15):
    n = input('enter a number:')
    if n.isnumeric():
        x.append(int(n))

total = sum(x)
avg = total/len(x)
print(f'sum => {total} and avg => {avg} of the list')
```

wap to add colors in a list that have character r in them

```
colors_with_r = []
for i in range(10):
    color = input('colorname:')
    if 'r' in color:
        colors_with_r.append(color)
    else:
        print('next->')
print(colors_with_r)
```

wap to write a program to store a fibonacci series a list

```
fib = [0,1]
for i in range(100):
    fib.append(fib[-1] + fib[-2])
print(fib)
```
