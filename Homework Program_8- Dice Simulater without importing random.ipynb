# WAP to create a dice simulater without using random 

```python
#@title Random Dice Simulation
import datetime
while True:
    present_time=list(datetime.datetime.now().strftime('%S'))
    le=len(present_time)
    for i in range(le):
        present_time[i]=int(present_time[i])
    if present_time[1]==0:
        present_time[1]+=1
    if present_time[1]>6:
        present_time[1]-=4
    inpt=input("Do you want to roll the dice? Print enter to stop.")
    if inpt:
        print(present_time[1])
    else:
        break
```
