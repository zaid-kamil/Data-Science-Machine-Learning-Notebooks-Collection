# lambda expressions






```
# single line expression 
# x + 10 * 3 / x
# AND
# you have a list of data
# [1,2,3,4,5,56,6,6,87,8,9,10]
# and that expression is to be applied for each value
```

```
# mapping
data =  [1,2,3,4,5,56,6,6,87,8,9,10] # iterator
list(map(lambda x: x + 10 * 3 /x, data))
```

```
# example feature scaling
data = [23,45,67,78,34,23,12,54,54,67,88,32] 
mx = max(data)
list(map(lambda i: i/mx, data))
```

```
import pandas as pd
```

```
import seaborn as sns
df = sns.load_dataset('penguins')
df
```

```
df.info()
```

```
# missing values
for col in df:
    print(df.isnull()[col].value_counts())
    print('---'*10)
```

## rules for handling missing values
- if the col type is numeric, replace the Nan with mean of col
- if col type is object, replace with mode(most common value)
- if col is selected as prediction col, them remove those rows with Nan

```
df['culmen_depth_mm'].mean()
```

```
df['culmen_depth_mm'].replace(np.nan, df['culmen_depth_mm'].mean(), inplace=True)
```

```
df['culmen_length_mm'].replace(np.nan, df['culmen_length_mm'].mean(), inplace=True)
df['flipper_length_mm'].replace(np.nan, df['flipper_length_mm'].mean(), inplace=True)
df['body_mass_g'].replace(np.nan, df['body_mass_g'].mean(), inplace=True)
```

```
# object
common_value = df['sex'].value_counts().idxmax() # first index value of output
df['sex'].replace(np.nan, common_value, inplace=True)
```
