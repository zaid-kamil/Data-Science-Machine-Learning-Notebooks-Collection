```
videogames=['Halo','Doom','Witcher3','Batman','Call of duty','Read Dead Redemption']
for game in videogames:
    print(game)
```

```
# enumerate example
for i, game in enumerate(videogames):
    print(i+1, game)
```

```
# enumerate example
for i, game in enumerate(videogames):
    if i % 2!=0:
        print(i, game)
```

```
ratings = [8,9,10,9,7,9]
```

```
# zip example
# zip allows us to pass multiple iterators into a single loop

for game,stars in zip(videogames, ratings):
    print(game,'has',stars,'rating')
```

```
data = list(range(1,101))
x = int(input('where do you want to stop: '))
for i in data:
    if i == x:
        break
    print(i, end=',')
```

```
# 5 or 7 divisble values should be skipped in loop

for i in data:
    if i%5==0 or i%7==0:
        continue
    print(i,end=' ')
```
